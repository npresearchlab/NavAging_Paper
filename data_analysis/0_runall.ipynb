```python
from IPython import get_ipython
import warnings
import pandas as pd

# Suppress pandas warnings
warnings.filterwarnings('ignore', category=pd.errors.SettingWithCopyWarning)
warnings.filterwarnings('ignore', category=pd.errors.PerformanceWarning)

get_ipython().magic('reset -sf')

# CHANGE THE FILE PATHS BELOW FOR YOUR COMPUTER

# Define the list of folder paths to process
fp_folders = [
    '/Volumes/YB_Drive/NavAging_Paper/data/YA_Data/',
    '/Volumes/YB_Drive/NavAging_Paper/data/OA_Data/'
]

fp_code = '/Volumes/YB_Drive/NavAging_Paper/data_analysis/' # filepath to codes

# Loop through each folder path
for i, fp_folder in enumerate(fp_folders):
    print(f"Processing folder {i+1}/{len(fp_folders)}: {fp_folder}")
    
    # Store the current folder path
    %store fp_folder
    
    # Run the analysis scripts
    get_ipython().run_line_magic('run', fp_code + '1_calculate_outcomes.ipynb')
    get_ipython().run_line_magic('run', fp_code + '2_merge_data.ipynb')
    #make sure to change save location at the end of script
    get_ipython().run_line_magic('run', fp_code + '3_average_data.ipynb')
    #make sure to change save location at the end of script
    get_ipython().run_line_magic('run', fp_code + '4_target_data.ipynb')
    get_ipython().run_line_magic('run', fp_code + '5_graph_data.ipynb')
    get_ipython().run_line_magic('run', fp_code + '6_post_analyses.ipynb')
    
    print(f"Completed processing for: {fp_folder}")
    print("-" * 50)

print("All processing complete!")
```
