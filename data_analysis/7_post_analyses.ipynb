```python
import pandas as pd
import os
```

```python
# Get the parent directory of fp_folder
csv_path = "/Volumes/YB_Drive/NavAging_Paper/data/oa_merged_results.csv"

# Read the CSV file
df = pd.read_csv(csv_path)

# Find the row with "BNC039" in the "Participant" column AND "Pawn Shop" in "Target_Name" column
condition = (df['Participant'] == 'BNC39') & (df['Block_num'] == 1) & (df['Target_Name'] == 'Pawn Shop')

# Before removing Orientation_Time, calculate new Total_Time by subtracting Orientation_Time from Total_Time
df.loc[condition, 'Total_Time'] = df.loc[condition, 'Total_Time'] - df.loc[condition, 'Orientation_Time']

# Now remove the Orientation_Time value
df.loc[condition, 'Orientation_Time'] = ''

# Save the modified dataframe back to the CSV file
df.to_csv(csv_path, index=False)

print("Successfully removed Orientation_Time value for participant BNC039 with Target_Name 'Pawn Shop'")
```
